```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
df = pd.read_csv('strong_3-1-20.csv')
```

```python
df[['Date', 'Workout_Name', 'Exercise_Name', 'Set_Order', 'Weight', 'Weight_Unit', 'Reps', 'Distance', 'Distance_Unit', 'Seconds', 'Notes', 'Workout_Notes']] = df.Date.str.split(';', expand=True,)
```

```python
df['Date'] = pd.to_datetime(df.Date)
df['Set_Order'] = pd.to_numeric(df.Set_Order)
df['Weight'] = pd.to_numeric(df.Weight)
df['Reps'] = pd.to_numeric(df.Reps)
```
